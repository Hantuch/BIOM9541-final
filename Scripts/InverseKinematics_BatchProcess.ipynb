```python
import opensim as osim
import os
import xml.etree.ElementTree as ET

osim.ModelVisualizer.addDirToGeometrySearchPaths("C:/OpenSim 4.5/Geometry")

base_path = "C:/Users/Hanif/OneDrive - UNSW/T3 - 2025/Biomechanics of Human Body/Final Project/Vicon Data/Shallow Squat"
model_path = "C:/Users/Hanif/OneDrive - UNSW/T3 - 2025/Biomechanics of Human Body/Final Project/Vicon Data/Models"
trc_path = os.path.join(base_path, 'Patched')

model_file = 'Hanif-Scaled.osim'

trc_files = [f for f in os.listdir(trc_path) if f.endswith('.trc')]

print(f"Loading model from:", model_file)
print(f"TRC files found {len(trc_files)}:", trc_files)
```

```python
# Inverse Dynamics Batch Processing
model = osim.Model(os.path.join(model_path, model_file))
model.initSystem()
for trc_file in trc_files:

    if not os.path.exists(os.path.join(base_path, 'IK', 'Continuous')):
        os.mkdir(os.path.join(base_path, 'IK', 'Continuous'))
    
    storage = osim.Storage(os.path.join(trc_path, trc_file))
    start_time = storage.getFirstTime()
    end_time = storage.getLastTime()

    ik_tool = osim.InverseKinematicsTool("./Setup_XMLs/IK_Setup.xml")
    ik_tool.setStartTime(start_time)
    ik_tool.setEndTime(end_time)
    ik_tool.setModel(model)
    ik_tool.setMarkerDataFileName(os.path.join(trc_path, trc_file))
    ik_tool.setOutputMotionFileName(os.path.join(base_path, 'IK', 'Continuous', trc_file.replace('.trc', '.mot')))
    ik_tool.run()

    print(f"Processing IK for: {trc_file}")
```
